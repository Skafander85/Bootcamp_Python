```python
#gra w oczko

# classy i definicje
import random

#elemnty z ktorych zbuduje talie
suits = ('Pik', 'Kier', 'Trefl', 'Karo')
ranks = ('Dwójka', 'Trójka', 'Czwrórka', 'Piątka', 'Szóstka', 'Siudemka', 'Ósemka', 'Dziewiątka', 'Dziesiątka', 'Walet', 'Królowa', 'Król', 'As')
values = {'Dwójka':2, 'Trójka':3, 'Czwrórka':4, 'Piątka':5, 'Szóstka':6, 'Siudemka':7, 'Ósemka':8, 'Dziewiątka':9, 'Dziesiątka':10, 'Walet':10,
         'Królowa':10, 'Król':10, 'As':11}

playing = True

#objekt katy
class Card:
    
    def __init__(self,suit,rank): #oczekuje stringa "Suit Rank"
        self.suit = suit
        self.rank = rank
    
    def __str__(self):
        return self.suit + " " + self.rank
    
    def rank(self): #zwracam rangę karty 
        
        return self.rank

#objekt talii, buduje, mieszam i wyciagam jedna

class Deck:
    
    def __init__(self):
        self.deck = []  # start with an empty list
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))  
                
    def __str__(self):
        deck_cały = ''  # start with an empty string
        for card in self.deck:
            deck_cały += '\n '+card.__str__() # add each Card object's print string
        return 'Talia kart:' + deck_cały
    

    def shuffle(self): #shufle
        random.shuffle(self.deck)
        
    def deal(self): #wypierdala pierwsza karte
        wylosowana_karta = self.deck.pop(0)
        return wylosowana_karta

#objekt ręki, mamm tam wartosci i sledze asy
class Hand:
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces
    
    def add_card(self,card):
        self.cards = self.cards + [card]
        self.value += values[card.rank]
        if card.rank == 'As':
            self.aces += 1  # #dodaje asy do puli
    
    def adjust_for_ace(self):
        
        if self.value >21 and self.aces: # jesli masz powyzej 21 i masz asa to obnizy ci wartosc o 10
            self.value -= 10   
            self.aces -= 1

#hajsy

class Chips:
    
    def __init__(self):
        self.total = 100  # This can be set to a default value or supplied by a user input
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
    
    def lose_bet(self):
        self.total -= self.bet
        
    def check_total(self):
        if self.total == 0:
            print("Dodaje 100 do twej póli bo się skończyło")
            self.total = 100
        

#stawka, sprawdzam czy nie za duzo stawia
def take_bet(chips):
    while True:
        try:
            chips.bet = int(input("Ile chcesz postawić?"))
        except:
            print("Podaj liczbę całkowitą!")
        else:
            if chips.bet > chips.total:
                print ("Masz za mało kasy! Biedaku!")
            else:
                break
        
#Dodawanie karty do reki - HIT, sprawdzam asy
def hit(deck,hand):
    hand.add_card(deck.deal())
    hand.adjust_for_ace()
    
#pytam czy dobiera czy stoi
def hit_or_stand(deck,hand):
    global playing  # to control an upcoming while loop
    
    while True:
        pytanie = input("Hit? T lub N:") #pytam czy dawać mu karty
        
        if pytanie[0].lower() == 't':
            hit(deck,hand)  # hit() function defined above
            
        elif pytanie[0].lower() == 'n': #nie dobiera i teraz kolej dilera
            print("Gracz stoi. Dealer teraz gra.")
            playing = False
            
        else:
            print("Podaj T lub N!")
            continue
            
        break
        
# wykładam karty na stol jedna ukrywam u dilera, lub pokazuje wszystkie
def show_some(player,dealer):
    print ('\nRęka dilera:')
    print ('\n <Ukryta karta>')
    print (f'\n {dealer.cards[1]}')
    print ('\nRęka Gracza:',*player.cards, sep="|")
    print(f"Wartość gracza={player.value}")

    
def show_all(player,dealer):
    print("\nRęka Dilera:", *dealer.cards, sep='\n ')
    print("Wartość Dilera =",dealer.value)
    print("\nRęka Gracza:", *player.cards, sep='\n ')
    print("Wartość Gracza =",player.value)

#funckje z możliwymi zakończeniami
def player_busts(player,dealer,chips):
    print("Przegrałeś!")
    chips.lose_bet()
    chips.check_total()

def player_wins(player,dealer,chips):
    print("Wygrałeś!")
    chips.win_bet()

def dealer_busts(player,dealer,chips):
    print("Wygrałeś!")
    chips.win_bet()
    
def dealer_wins(player,dealer,chips):
    print("Diler Wygrywa!")
    chips.lose_bet()
    chips.check_total()
    
def push(player,dealer):
    print("Remis! It's a push.")
    
player_chips = Chips()
#---------------------============================================-----------------------------
#tu zaczyna się gra:
while True:
    # Print an opening statement
    print('Gramy w Black Jacka, chuje!')

    
    # Create & shuffle the deck, deal two cards to each player
    talia = Deck()
    talia.shuffle()
    
    player = Hand()
    player.add_card(talia.deal())
    player.add_card(talia.deal())
    
    dealer = Hand()
    dealer.add_card(talia.deal())
    dealer.add_card(talia.deal())
        
    # Set up the Player's chips
    #player_chips = Chips()
    print (f"\n Masz tyle: {player_chips.total}")
    # Prompt the Player for their bet
    
    take_bet(player_chips)
    
    # Show cards (but keep one dealer card hidden)
    show_some(player, dealer)
    
    
    while playing:  # recall this variable from our hit_or_stand function
        
        # Prompt for Player to Hit or Stand
        hit_or_stand(talia, player)
        
        # Show cards (but keep one dealer card hidden)
        show_some(player, dealer)
        print (f"\n pynkty gracza: {player.value}")
        
        # If player's hand exceeds 21, run player_busts() and break out of loop
        if player.value > 21: 
            player_busts(player,dealer,player_chips)
            break

    # If Player hasn't busted, play Dealer's hand until Dealer reaches 17
    if player.value <= 21:
        while dealer.value < 17: #dobieraj mu aż wyjdzie poza 17
            hit(talia,dealer)
            
        
        # Show all cards
        show_all(player, dealer)
        
        # Run different winning scenarios
        if dealer.value > 21:
            dealer_busts(player, dealer, player_chips)
            
        elif dealer.value > player.value:
            dealer_wins(player, dealer, player_chips)
            
        elif dealer.value < player.value:
            player_wins(player, dealer, player_chips)
        
        else:
            push(player,dealer)
    
    # Inform Player of their chips total 
    print (f"Twoja pula: {player_chips.total}")
    
    # Ask to play again
    grasz = input("Grasz jeszcze raz? t lub n")
    if grasz[0].lower() == "t":
        playing = True
        continue
    else:
        print('No to wypierdalaj!')
        break
```
